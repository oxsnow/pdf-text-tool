```python
import csv
import os
```

```python
def extract_ticker_and_year(filename):
    # Split the filename into parts
    parts = filename.split('_')
    if len(parts) == 3:
        ticker = parts[0]
        year = parts[2]
    else:
        ticker = filename
        year = "-"
    return ticker, year
```

```python

def verify_file_paths(file_paths, csv_file_path):
    """
    Verifies that all file paths in `file_paths` are included in the CSV file.

    :param file_paths: List of file paths to verify.
    :param csv_file_path: Path to the CSV file containing ticker and year columns.
    :return: A tuple of (missing_files, extra_files).
    """
    # Extract the file names from the file_paths
    file_names_path = [os.path.basename(path) for path in file_paths]
    file_names_to_check = []
    for file in file_names_path:
        ticker, year = extract_ticker_and_year(file)
        file_names_to_check.append(f"{ticker} {year}")

    # Read the CSV file and construct the list of file names
    file_names_from_csv = []
    with open(csv_file_path, "r", encoding="utf-8") as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            # Combine 'ticker' and 'year' to generate the file name
            ticker = row["Ticker"]
            year = row["Year"]
            file_name = f"{ticker} {year}"  # Adjust the format if needed
            file_names_from_csv.append(file_name)

    # Check for missing and extra files
    missing_files = [file for file in file_names_to_check if file not in file_names_from_csv]
    extra_files = [file for file in file_names_from_csv if file not in file_names_to_check]

    return missing_files, extra_files
```

```python
pdf_files = []
    
for file_name in os.listdir('D:\\Hp\\OneDrive\\shared\\annual report renamed\\sustainability\\2023'):
    if file_name.endswith(".pdf"):  # Check if the file is a PDF
        full_path = os.path.join('D:\\Hp\\OneDrive\\shared\\annual report renamed\\sustainability\\2023', file_name)
        file_names_without_extension = full_path.split('.')[0]
        pdf_files.append(file_names_without_extension)

pdf_files
```

```python


# Example usage
csv_file_path = "D:\\Hp\\OneDrive\\project\\pdf-text-tool\\pdf_text_tool\\build\\windows\\x64\\runner\\Release\\release-1.2.3\\summary-2025-04-28 21-11-36.csv"

missing, extra = verify_file_paths(pdf_files, csv_file_path)

print("Missing files:", missing)
print("Extra files in CSV:", extra)
```
